```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pitches = pd.read_csv("../pitches.csv")
atbats = pd.read_csv("../atbats.csv")
games = pd.read_csv("../games.csv")
```

```python
atbats = atbats.loc[:, ['g_id', 'ab_id', 'top']]
```

```python
atbats.head()
```

```python
games = games.loc[:, ['away_team', 'g_id', 'home_team']]
games.head()
```

```python
pitches = pitches.loc[:, ['px', 'pz', 'code', 'pitch_type', 'ab_id']]
```

```python
# df.set_index('key').join(other.set_index('key'))
pitches_atbats = pd.merge(pitches, atbats)
```

```python
all_data = pd.merge(pitches_atbats, games)
all_data.head()
```

```python
all_data.tail()
```

```python
all_data.to_csv('../alldata.csv')
```
